```python
with open('D:\CPLEX-ORLAB\large_scale\knapPI_3_10000_1000_1', 'r' , encoding= 'utf-8' ) as f :
    text = f.read()
text = text.split('\n')
```

```python
for x in range(len(text)):
    text[x] = text[x].split()
capacity = float(text[0][1])
text = text[1:-2]
print(text)
```

```python
import numpy as np
from docplex.mp.model import Model
```

```python
opt_mod = Model(name = 'Knapsack')
n = len(text)
x = np.zeros(n)
print(x)
```

```python
x = [opt_mod.binary_var(name = f'x{i}') for i in range(n)]
x
```

```python
opt_mod.set_objective('max' , sum([x[i] * float(text[i][0]) for i in range(n)]))
```

```python
opt_mod.add_constraint( sum([x[i] * float(text[i][1]) for i in range(n)]) <= capacity)
```

```python
import time 
start_time = time.time()
print( opt_mod.solve())
print(f'Thoi gian chay: {time.time() - start_time} ')
```
