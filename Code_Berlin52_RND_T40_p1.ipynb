```python
with open('11berlin52_RND_T40_p1.sop' ,'r',encoding = 'utf-8') as f :
    text = f.read()
print(text)
```

```python
import re 
pattern = r'(\d+)\s+([\d]+\.[\d]+)\s+([\d]+\.[\d]+)'
matches = re.findall(pattern,text)
```

```python
import numpy as np
c = np.zeros((52,52))
```

```python
import math
for i in range(52):
    for j in range(52):
        c[i][j] = math.sqrt( (float(matches[i][1]) - float(matches[j][1]))**2 + (float(matches[i][2]) - float(matches[j][2]))**2 )
```

```python
from docplex.mp.model import Model
```

```python
from docplex.mp.model import Model

# Khởi tạo mô hình
opt_mod = Model(name='TSP')

n = 52
x = [[opt_mod.binary_var(name=f'x_{i}_{j}') for j in range(n)] for i in range(n)]

```

```python
#Tong cua x[i][j] == 1; i =1,n
#Voi moi gia tri i thi ta them 1 rang buoc
#for i in range(n):
    #opt_mod.add_constraint(sum([x[i][j] for j in range(n) if i != j] ) ==1)
for i in range(n):
    ct = []
    for j in range(n):
        if i != j:
            ct.append(x[i][j])
    opt_mod.add_constraint(sum(ct) == 1) 
```

```python
#tong cua x[i][j] == 1 ; j=1,n
#voi moi gia trị j thì ta thêm 1 ràng buộc
for j in range(n):
    opt_mod.add_constraint(sum([x[i][j] for i in range(n) if i != j] ) == 1 )
```

```python
ob_func = opt_mod.set_objective('min' , sum( [x[i][j] * c[i][j] for i in range(n) for j in range(n) if i != j]))
```

```python
#u denotes cho thứ tự đến 
#u bang bao nhieu 0 quan trong , Vi du u2 = 2 < u7 = 3 thi den 2 truoc 7
#vi du u2 = 2 , u7 = 3 thi di tu 2-->7
#u=2 thi den thanh pho day' dau` tien xong cu tiep tuc
u = [0]
for i in range(1,n):
    u += [opt_mod.integer_var(name = f'u{i}', lb =1 , ub = n)]
```

```python
for i in range(1,n):
    for j in range(1,n):
        if i != j :
            opt_mod.add_constraint(u[i] - u[j] + 1 <= (n-1)*(1-x[i][j]))

```

```python
import time
start_time = time.time()
(opt_mod.solve())
end_time = time.time()
print(opt_mod.solve())
print(f'Thoi gian chay {end_time - start_time}')
```
